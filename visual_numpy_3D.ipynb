```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(0,101):

    # Crear un array tridimensional con valores entre -100 y 100
    shape = (10, 10, 10)  # Dimensiones del array
    array_3d = np.random.randint(-100, 101, size=shape)

    # Para representar gr√°ficamente el array tridimensional,
    # tomaremos una suma proyectada a lo largo de un eje o una rebanada
    projected_2d = array_3d.sum(axis=2)  # Proyecci√≥n sumada a lo largo del eje Z

    # Crear el gr√°fico
    plt.figure(figsize=(8, 6))
    plt.imshow(projected_2d, cmap='coolwarm', aspect='auto')
    plt.colorbar(label='Valor proyectado')
    plt.title(f'Proyecci√≥n 2D de un Array 3D - {i}')
    plt.xlabel('Eje X')
    plt.ylabel('Eje Y')

    # Exportar el gr√°fico
    output_file = f"{i}_3d_array_projection.png"
    plt.savefig(output_file)
    plt.show()

    print(f"Gr√°fico exportado como {output_file}")
```
